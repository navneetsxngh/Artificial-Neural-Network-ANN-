```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Data\Attrition - Attrition.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
cat_cols = [x for x in df.columns if df[x].dtype == 'object']
cat_cols = ['BusinessTravel',
 'Department',
 'EducationField',
 'Gender',
 'JobRole',
 'MaritalStatus',
 'Over18',
 'OverTime']
cat_cols
```

```python
X = df.drop(columns=['Attrition'], axis=1)
Y = df['Attrition']
```

```python
X_encoded = pd.get_dummies(X, columns= cat_cols, drop_first=True)
X_encoded = X_encoded.astype(int)
X_encoded.head()
```

```python
X_encoded.info()
```

```python
lb = LabelEncoder()
Y_encoded = lb.fit_transform(Y)
Y_encoded
```

```python
X_Train, X_test, y_train, y_test = train_test_split(X_encoded, Y_encoded, test_size=0.2, random_state=42)
```

```python
X_Train.head()
```

```python
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_Train)
X_test_scaled = sc.transform(X_test)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential([
    Dense(units=128, activation='relu', input_shape=(X_train_scaled.shape[1],)),     ## Input Layer
    Dropout(0.3),
    Dense(units=64, activation='relu'),                                        ## Hidden Layer
    Dropout(0.3),
    Dense(units=1, activation='sigmoid')                                       ## Output layer
])
```

```python
model.summary()
```

```python
loss_fn = tf.keras.losses.BinaryCrossentropy()
opt = tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.9, nesterov=True)
```

```python
model.compile(optimizer=opt, loss=loss_fn, metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
## Training the Model
model.fit(
    X_train_scaled, y_train, validation_data=(X_test_scaled, y_test), epochs=50, batch_size = 8, validation_split=0.1,)
```

```python
loss, accuracy = model.evaluate(X_test_scaled, y_test)
```

```python
print(f"Test Loss : {loss:.4f} Test Accuracy = {accuracy:.4f}")
```
